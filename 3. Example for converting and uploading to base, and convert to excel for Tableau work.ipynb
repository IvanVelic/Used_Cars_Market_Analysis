```python
#ukloni duple redove, ali zadrzi prvog
data = data.drop_duplicates(keep='first')

#ukloni redove koji nemaju podatak iz kolone Price
data = data.dropna(subset=['Price(€)'])

# Drop rows with "- " in the "Year" column
data = data.drop(data[data['Year of production'].str.contains('\-')].index)
data = data.drop(data[data['Power(hp)'].str.contains('\-')].index)

#remove rows where the "Fuel type" is exactly a hyphen (-) and not part of any other value. Values like 'Hybrid - Benzin' will not be affected
data = data.drop(data[data['Fuel type'] == '-'].index)

# Drop rows with "- " in the "Gearbox" column
data = data.drop(data[data['Gearbox'].str.contains('\-')].index)
data = data.drop(data[data['Mileage(km)'].str.contains('\-')].index)
data = data.drop(data[data['Fuel consumption (l/100km - comb.)'].str.contains('\-')].index)
data = data.drop(data[data['CO2 emissions (g CO2/km - comb.)'].str.contains('\-')].index)
#Can only use .str accessor with string values!
#data = data.drop(data[data['Price(€)'].str.contains('\-')].index)
data = data.drop(data[data['Location'].str.contains('\-')].index)

print("Number of datapoints:", len(data))
display(data.head(10))  
display(data.tail(10)) 
```

```python
import datetime

data = pd.read_csv('C:/Users/Ivan/Desktop/Hey.car/ALL/1.clean.csv')

data['Year of production'] = pd.to_datetime(data['Year of production'], format='%m/%y')

current_year = datetime.datetime.now().year  # Define current_year before using it in the lambda function

# Adjust year values based on the given conditions
data['Year of production'] = data['Year of production'].apply(lambda x: x.replace(year=x.year + 100) if x.year < current_year - 2000 else x.replace(year=x.year))

data['Year of production'] = data['Year of production'].dt.strftime('%Y-%m-%d')

data['Age(years)'] = current_year - pd.DatetimeIndex(data['Year of production']).year

data['Age(years)'] = data['Age(years)'].astype(int)

# Reorder columns with Years in use
column_order = ['Model', 'Edition', 'Year of production', 'Age(years)', 'Power(hp)', 'Fuel type', 'Gearbox', 'Mileage(km)',
                'Fuel consumption (l/100km - comb.)', 'CO2 emissions (g CO2/km - comb.)', 'Price(€)', 'Location']
data = data[column_order]

print("Number of datapoints:", len(data))
display(data.head(10))
display(data.tail(10))
```

```python
import datetime

data = pd.read_csv('C:/Users/Ivan/Desktop/Hey.car/ALL/2.add_age.csv')

# Convert columns to appropriate data types and apply transformations

data['Model'] = data['Model'].astype(str)
data['Edition'] = data['Edition'].astype(str)
data['Year of production'] = pd.to_datetime(data['Year of production'], format='%Y-%m-%d')
data['Age(years)'] = datetime.datetime.now().year - pd.DatetimeIndex(data['Year of production']).year
data['Age(years)'] = data['Age(years)'].astype(int)

data['Power(hp)'] = data['Power(hp)'].astype(str).str.extract('(\d+)', expand=False).astype(int)

fuel_type_mapping = {
    'Elektro': 'Electric',
    'Benzin': 'Gasoline',
    'Hybrid - Benzin': 'Hybrid/Gasoline',
    'Diesel': 'Diesel',
    'Autogas - LPG': 'LPG',
    'Gas': 'CNG',
    'Andere': 'Others',
    '-': '-',
    'Hybrid - Diesel': 'Hybrid/Diesel'
}
data['Fuel type'] = data['Fuel type'].replace(fuel_type_mapping)
data = data[data['Fuel type'] != '-']  # Drop rows with only hyphen

gearbox_mapping = {
    'Automatik': 'Automatic',
    'Schaltgetriebe': 'Manual'
}
data['Gearbox'] = data['Gearbox'].replace(gearbox_mapping)

data['Mileage(km)'] = data['Mileage(km)'].astype(str).str.replace('[^\d]', '', regex=True).astype(int)

data['Fuel consumption (l/100km - comb.)'] = data['Fuel consumption (l/100km - comb.)'].str.extract('([\d,\.]+)', expand=False).replace(',', '.', regex=True).astype(float)

data['CO2 emissions (g CO2/km - comb.)'] = data['CO2 emissions (g CO2/km - comb.)'].str.extract('(\d+)', expand=False).astype(int)

data['Price(€)'] = data['Price(€)'].astype(float).astype(int)

data['Location'] = data['Location'].astype(str)

# Reorder columns
column_order = ['Model', 'Edition', 'Year of production', 'Age(years)', 'Power(hp)', 'Fuel type', 'Gearbox', 'Mileage(km)',
                'Fuel consumption (l/100km - comb.)', 'CO2 emissions (g CO2/km - comb.)', 'Price(€)', 'Location']
data = data[column_order]

print("Number of datapoints:", len(data))
display(data.head(10))
display(data.tail(10))
```
